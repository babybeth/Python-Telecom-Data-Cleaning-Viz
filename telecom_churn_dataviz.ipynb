Data Visualization for Telecom data in Python

The data file used for visualization purpose is telecom_churn_cleaned.csv file. Here data visualization is done using matplotlib and seaborn libraries.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

####Find the most frequent complaint code

```python
sns.countplot(data=telecom_data,x='MOST_FREQ_COMPLAINT_CD')
```

The plot  shows a linear relationnship bretween total email queries last month and the number of services dropped to some extent. But as number of services dropped increases, it is not well explained by the email queries. This means that either the complaints are not all recorded or was submitted in a different way which is not included in the dataset.

```python
sns.lineplot(data=telecom_data, x='TOT_SRV_DROPPED_CNT',y= 'TOT_EMAIL_QUERY_CNT')
```

From the histotogram, majority of email queries are sent by those customers that has a occupation code of 1 to 10.So need to check if there is anything common to those with this occupation code

```python
sns.histplot(data=telecom_data, x= 'TOT_EMAIL_QUERY_CNT',hue='OCCUPATION_CAT_CD')
```

The joint plot shows that the email queries are sent by both males and females follows the same pattern.

```python
sns.jointplot(data=telecom_data,x='OCCUPATION_CAT_CD',y='TOT_EMAIL_QUERY_CNT',hue='GENDER_CD')
```

Majority of customers in this dataset has occupation code 10.This will give an insight on who used the email queries the most.

```python
plt.figure(figsize = (25,10))
sns.countplot(data=telecom_data,x='OCCUPATION_CAT_CD')
```

The kernel density distribution of tenure in network.It is normally distributed.

```python
sns.FacetGrid(telecom_data,hue='GENDER_CD',height=5).map(sns.kdeplot,'SUBS_TENURE').add_legend()
```

```python
sns.histplot(telecom_data['SUBS_TENURE'],kde = True)
```

```python
sns.distplot(telecom_data['SUBS_TENURE'],bins=10,kde=True)
```

The histogram for customer age shows that there are more number of customers with ages ranging from 22 to 28.

```python
sns.jointplot(data=telecom_data,x='CUST_AGE',y='TOT_ACTV_SRV_CNT',hue='GENDER_CD')
```

```python
sns.histplot(telecom_data['CUST_AGE'])
```

Plotting the number of customers wrt their ages and their customer segment code

```python
plt.figure(figsize=(15,5))
sns.countplot(telecom_data,x='CUST_AGE',hue='CUST_SEG_CD')
```

Plotting the number of customers in terms of their age who got a significant percentage change of > 2% and their customer segment code. The plot shows that customers belonging to age group > 45 got a better percentage change in the bill amount.

```python
plt.figure(figsize=(15,5))
sns.countplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT']  >2)],x='CUST_AGE',hue='CUST_SEG_CD')
```

The pairplot plots call duration against percentage bill change for local and national calls

```python
sns.pairplot(telecom_data[telecom_data['PCT_CHNG_BILL_AMT'] >2 ],vars=['PCT_CHNG_BILL_AMT','OB_CALL_LOC_DUR_BASE_5','OB_CALL_LOC_DUR_BASE_6','OB_CALL_NAT_DUR_BASE_5','OB_CALL_NAT_DUR_BASE_6'],kind='scatter',diag_kind='kde',hue='GENDER_CD')
```

Scatter plot for percentage change in bill amount wrt international call duration for base 5 month.

```python

sns.scatterplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT'] >2 )],x='OB_CALL_INTL_DUR_BASE_5',y='PCT_CHNG_BILL_AMT')
```

Scatter plot for percentage change in bill amount wrt international call duration for base 6 month.

```python
sns.scatterplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT'] >2 )],x='OB_CALL_INTL_DUR_BASE_6',y='PCT_CHNG_BILL_AMT')
```

These scatter plots shows the percentage change in bill amount relative to local call duration in base months 5 and 6

```python
sns.relplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT'] >0 )],x='OB_CALL_LOC_DUR_BASE_5',y='PCT_CHNG_BILL_AMT',color='orange',kind = 'scatter',hue='CURR_PRICE_PLAN_CD')
sns.relplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT'] >0 )],x='OB_CALL_LOC_DUR_BASE_6',y='PCT_CHNG_BILL_AMT',color='purple',kind = 'scatter',hue='CURR_PRICE_PLAN_CD')
plt.show()
```

This plot shows that those with better percentage change in bill amount use an average of 3 active services. There are not many with a great percentage change has more than 4 active services.So no:of active srvices has an effect on change in bill amount.

```python
sns.scatterplot(telecom_data[(telecom_data['PCT_CHNG_BILL_AMT'] >2 )],x='TOT_ACTV_SRV_CNT',y='PCT_CHNG_BILL_AMT',hue='CUST_AGE')
```

This plot shows that people belonging to less than 22 yrs use less usage of active services and this may be attributed to less active customers in that age group. While most males in this age group use between 0 to 2 services, those females who use more than 3 services outnumber males using the same.

```python
plt.figure(figsize=(12,10)) 
sns.boxplot(telecom_data,x='CUST_AGE',y='TOT_ACTV_SRV_CNT',hue='GENDER_CD')
```

The peak in the histogram for male in the TOT_ACTV_SRV_CNT shows male dominance in using active subscribers at age 20 which is more clear from the next plot

```python
df= telecom_data.loc[telecom_data['CUST_AGE'].isin([20])].loc[:,['GENDER_CD','TOT_ACTV_SRV_CNT','TOT_SRV_ADDED_CNT','TOT_SRV_DROPPED_CNT']]
sns.pairplot(df,hue='GENDER_CD')
```

This plot shows in males using more active services in average at age 20

```python
sns.countplot(df,x='TOT_ACTV_SRV_CNT',hue='GENDER_CD')
```
