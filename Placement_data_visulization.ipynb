```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_csv('/content/Placement_Data.csv')
```

```python
df.head()
```

```python
#Check the shape of the data
df[df['status']=='Placed'].shape
```

```python
#Statistical summary of the data
df.describe()
```

```python
#Distribution of eligilbility test
plt.figure(figsize=(10,4))
sns.histplot(data=df,x='etest_p', kde = 'True')
```

```python
#bargraph showing the count of students in each specialization based on teir employment status
sns.countplot(data=df,x='hsc_s',hue = 'status')
```

Commerce Students has the highest number of placements and Arts have the least.
In Arts, placed and not placed students are almost same.

```python
#Student's specialization in higher secondary education
sns.countplot(data=df,x='hsc_s',hue='hsc_s')
```

Commerce is the most preferred specialization among students.
Arts is the least preferred.
95% of students prefer to join in Commerece or Arts.

```python
#Corelation between the various percentages
correlation = df.corr()
correlation
```

```python
sns.heatmap(correlation,annot=True,cmap='YlGnBu')
```

Degree and Secondary school percentage are most correlated.
There is not much education percentages and employability test percentage.
Those with good percentage in degree have more chance of getting good percentages in MBA.

```python
#Visualization of student's degree percentage and their specialization
sns.boxplot(data=df,x='degree_p',y='hsc_s')
```

There are no outliers in data available for Commerce and Arts specialization.
For Science specialization, data is right skewed and there are outliers present.
75% of Commerce  and Science students have degree percentage less than 73%.
75% of Arts students have degree percentage less tha 68%.


```python
#Visualization for secondary education percentage and higher secondary education percentage based on the employment status
sns.catplot(data=df,x='ssc_p',y='hsc_p',kind = 'point',hue='status')
```

```python
#Best fit line for place and not placed students
sns.lmplot(data=df,x='ssc_p',y='hsc_p',hue='status')
```

The best fit line for placed students lie above the not placed students.

```python
#Distribution of Degree percentage for students based on employment status
sns.histplot(data=df,x='degree_p',hue = 'status',kde = True)
```

```python
#Boxplot visualization of secondary education percentage for placed and not placed students
sns.boxplot(data=df,x='ssc_p',y='status')
```

75% of not placed students have their secondary education percentage less than 63%

```python
#Degree percentage of all the disciplines categorized for placed and not placed students
plt.figure(figsize=(15,7))
sns.boxplot(data=df,y='degree_p',x='degree_t',hue='status')
```

50% 0f students that are placed have a higher degree percentage than not placed students.For Comm&Mgmt, 75% of placed students have higher degree than not placed students.

```python
#Pairplot for degree percentage and secondary education percentage
sns.jointplot(data=df,x='ssc_p',y='degree_p',kind='hex')
```

Students who score between 60-70% in ssc are more likely to score in the range of 60-65% in degree.
Secondary Education percentage distribution is almost a normal distribution.

```python
#Distribution of employablity test percentage
sns.histplot(data=df,x='etest_p',kde=True)
```

```python
df['etest_p'].max()
```

Employability test percentage data is a right-skewed distribution. Highest percentage scored in the Employability test percentage is 98%.

```python
#Violin plot of students with specialization and their degree percentage.
sns.catplot(data=df,x='degree_p',y='specialisation',kind='violin')
```

75% of students who opted for Marketing & HR and Marketing & Finance have their degree percentage below 80%
